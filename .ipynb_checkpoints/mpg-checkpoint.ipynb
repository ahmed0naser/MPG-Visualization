```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/user/Desktop/datasets/auto-mpg.csv")
```

# let's get rid of the column we won't need

```python
df.drop(columns='car name',inplace=True)
```

```python
df['model year'].unique()
```

```python
df.horsepower.unique()
```

```python
q=df[df.horsepower=='?']
q
```

# only 6 rows have no horsepower so we'll drop them

```python
df.drop(index=df[df.horsepower=='?'].index,inplace=True)
```

```python
df.shape
```

```python
q=df[df.horsepower=='?']
q
```

```python
df.horsepower=df.horsepower.astype('int16')
```

```python
df.info()
```

# origin is the origin country so we'll change the values to the names

```python
df.origin=df['origin'].replace({1:'USA',2:'Europe',3:'Asia'})
df.origin=df.origin.astype('category')
df.head()
```

```python
df.displacement.unique()
```

# no need for the displacment to be float

```python
df.displacement=df.displacement.astype('int64')
print(df.info())
print(df['model year'].unique())
```

```python
df.displacement
```

```python
df.info()
```

```python
sns.countplot(data=df,x='origin')
plt.show()
```

```python
sns.barplot(data=df,x='origin',y=df.horsepower,ci=None)
plt.show()
```

```python
sns.barplot(data=df,x='origin',y='weight',ci=None)
plt.show()
```

# so USA make the most and the heaviest cars let's see the correlation between that and the mpg

```python
sns.barplot(data=df,x='origin',y='mpg',ci=None)
plt.title('mpg against origin')
plt.show()
```

# as expected USA has the lowest miles per gallon as it's the heaviest cars and need the highest horsepower

```python
sns.scatterplot(data=df,x='weight',y='mpg')
plt.title('weight against miles per gallon')
plt.show()
```

# so there's negative correlation between weight and mpg which makes USA cars are the lowest mpg as resault of their heavy cars and also haves the highest horsepower to work with this heavy cars

```python
sns.barplot(data=df,x='model year',y='weight',ci=None,color='#7fa754')
plt.title('weight over the years')
plt.show()
```

```python
sns.pointplot(data=df,y=df.horsepower,x=df.cylinders,ci=None)
plt.xticks(rotation=90)
plt.show()
```

# sounds like there's no obvious correlation between cylinders and horsepower

```python
sns.barplot(data=df,x=df.cylinders,y=df.mpg,ci=None)
plt.show()
```

```python
sns.countplot(data=df,x=df.cylinders)
plt.show()
```

# so when cylinders are more than 4 mpg starts to degrease and most cars have 4 cylinders

# Finally let's see over the years how mpg changes in respect to origin countries

```python
sns.pointplot(x='model year',data=df,y='mpg',hue='origin',ci=None)
plt.title('mpg over the years ')
plt.show()
```

# so over the years mpg has increased but yet USA cars are the worst always when it comes to weight and mpg while europe and asia have been competing
